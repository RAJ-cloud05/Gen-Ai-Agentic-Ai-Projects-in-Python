PYTHON MATH MODULE

=> To perform some advanced operations on the numbers we need a module name as "math" module.
=> To to use the advanced operations we have to call the module otherwise we get "NameError".
=> To call the module we use a keyword "import" or "from modulename import ...".

```python
sqrt(25)     #this is the error when we dont call the module and perform the operations.
```
