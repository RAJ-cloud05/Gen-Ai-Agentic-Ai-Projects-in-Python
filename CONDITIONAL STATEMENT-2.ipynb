```python
i = 1

while i<=5:
    print('data science') # when we mention end then new line will not create
    j = 1
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()    
```

```python
i = 1
while i<=5:
    print(' datascience', end = "") # when we mention end then new line will not create
    j = 1
    while j<=4:
        print(' technology', end= "")
        j = j + 1
        
    i = i + 1
    print()
```

```python
i = 1

while i <= 2 :
    j = 0
    
    while  j <= 2 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```
