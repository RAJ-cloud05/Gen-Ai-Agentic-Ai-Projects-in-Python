PYTHON INDEXING

```python
s="WELCOME TO GEN AI AND AGENTIC AI"
```

```python
s[0]    #forward indexing
```

```python
s[1]      #forward indexing
```

```python
len(s)        #forward indexing
```

```python
s[-1]        #backward indexing
```

```python
s[-5]         #backward indexing
```

```python
s[-len(s)]      #backward indexing using len() inbuild function
```

PYTHON SLICING 

```python
a="WELCOME TO GEN AI AND AGENTIC AI"
```

```python
len(a)
```

```python
a[0:]   #starts from 0th index and include till end index value 
```

```python
a[:32]   #starts from the beginning index value and end with {END-1}th index 
```

```python
a[0:32]    #starts from the given index value and ends with {END-1}th index
```

```python
a[:]     #starts from 0th index and includes all the values till the last index value 
```

```python
a[0:32:1]       #starts from 0th index to {END-1}th index with a forward direction of steps=+1
```

```python
a[10:]
```

```python
a[25:]
```

```python
a[:25]
```

```python
a[:15]
```

```python
a[15:30]
```

```python
a[10:19]
```

```python
a[:]
```

```python
a[::-1]
```

```python
a[10:20:2]     #steps=+2
```

```python
a[1:31:9]     #steps=+9
```

```python
a[:19:-2]      #steps=-2
```

```python
a[3::-1]         #steps=-4
```
