PYTHON LIST DATASTRUCTURE

```python
l=[]    #this is treated as empty list
```

```python
l
```

```python
l=[10,20,30.5, "RAJ", "DATA SCIENCE", True, 2+3j]
```

```python
l
```

```python
l.append(100)    #append function is used to add the element at the last position of the list.
l
```

```python
l.count(10)    #count function is used to count the repeated terms in the list
```

```python
l2=[]
```

```python
l2=l.copy()      #this is shallow copy... memory id would be different in both of the list and when we modify one list another list wont be affected.
l2
```

```python
l2=l      #this is deep copy ..... memory id is same for both the list and the modification in one list would affect the other list.
l2
```

```python
l
```

```python
l2
```

```python
l.extend(l2) # extend function is used to add the one list members to another list members.
l
```

```python
l.index("RAJ")   #index function is used to return the index value of the particular element inside the list.
```

```python
l
```

```python
l.clear()     #clear function is  used to clear all the elements from the list
l
```
