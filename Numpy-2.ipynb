```python
x=np.random.randint(10,50,(10,10))
x
```

```python
x[3:8]   #displays the rows from 3rd ---> 8-1=7th 
```

```python
print(x[3,8])   #if we give "," in between then it will give the specific position value at 3rd row and 8th column
```

```python
print(x[-5])
```

```python
print(x[-5,-3])
```

```python
print(x[::3])   #+3 steps
```

```python
x[:,3]    # prints the specific column
```

```python
print(x[2])    #prints the 2nd row
print(x[:,5])    #prints the 5th column       
```

```python
print(x)
```

```python
print(x[2:5,4:7])    #this will print the values between 2-->4th row and 4-->6th column
```

# NUMPY MASKING OR FILTER

```python
z= np.arange(0,100).reshape(10,10)
z
```

```python
z < 50    #this will check the condition is the value less than 50 if yes then TRUE else FALSE
```

```python
z[z<50]    #list of values which are less than 50, applied both masking and filtering 
```

```python
z[z!=50]   #prints the values which is not equal to 50 
```
