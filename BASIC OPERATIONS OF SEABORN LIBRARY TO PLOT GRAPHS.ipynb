```python
df['InternetUsers']
```

```python
v= sns.distplot(df['InternetUsers'])     #with distribution line 
plt.show()
```

```python
v2= sns.displot(df['InternetUsers'])    #without distribution line 
```

```python
v3= sns.distplot(df['InternetUsers'], bins=10)  #here bins means the number of blue color boxes 
```

```python
v4= sns.boxplot(data=df, x='IncomeGroup', y='BirthRate')    #bi-variate analysis x&y
```

```python
v5= sns.lmplot(data=df, x='InternetUsers', y='BirthRate')    #linearModel plot   one value of x and one value of y is plotted in this plot
```

```python
v6= sns.lmplot(data=df, x='InternetUsers', y='BirthRate',fit_reg=False) #removes the regression line from the plot
```

```python
v7= sns.lmplot(data=df, x='InternetUsers', y='BirthRate',fit_reg=False, hue=  'IncomeGroup')      #hue==parameters for the color
```

```python
v7= sns.lmplot(data=df, x='InternetUsers', y='BirthRate',fit_reg=True, hue=  'IncomeGroup') 
```
