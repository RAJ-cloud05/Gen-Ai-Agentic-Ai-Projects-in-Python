```python
np.arange(10)    #same as range datatype in python basics
```

```python
np.arange(10,20)    #start and end-1
```

```python
np.arange(10,20,3)    #start and end-1 with steps=+3
```

```python
np.zeros(5)    # bydefault the parameter is this zeros(shape, dtype=float, order='C', *, like=None) & 1-D array
```

```python
np.zeros(5, dtype=int)  #changing the type from float to int   & 1-D array
```

```python
np.zeros((4,2))    # 2-D array and float datatype bydefault
```

```python
np.zeros((4,2), dtype=int)  #2-D array and int datatype  or 2x2 matrix
```

```python
np.zeros((5,10), dtype=int)    # 5x10 matrix
```

```python
np.ones(5)   #creates a array of only 1's value & its a 1-D value
```

```python
np.ones((3,5), dtype=int)   # 3x5 matrix with int datatype
```

```python
np.random.rand(4)   #generates random values and when we run again and again the values get shuffled
```

```python
np.random.randint(0,10)  #its returns any random value between start to end-1 any random value
```
